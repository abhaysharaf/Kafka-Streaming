```python
import uuid
import json
from confluent_kafka import Consumer
from google.colab import userdata

# 1. Securely fetch secrets from Colab's environment manager
try:
    ca_cert_content = userdata.get('KAFKA_CA')
    kafka_user = userdata.get('KAFKA_USER')
    kafka_password = userdata.get('KAFKA_PASSWORD')
except Exception as e:
    raise ValueError("Make sure KAFKA_CA, KAFKA_USER, and KAFKA_PASSWORD are set in your Colab Secrets (🔑 sidebar)!") from e

# 2. Write the certificate to a temporary local file
with open("ca.pem", "w") as f:
    f.write(ca_cert_content.strip())

# 3. Secure Configuration using variables
conf = {
    'bootstrap.servers': 'kafka-36fd6d61-abhaysharaf-512f.b.aivencloud.com:11895',
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': kafka_user,
    'sasl.password': kafka_password,
    'group.id': f'group-{uuid.uuid4()}',
    'auto.offset.reset': 'earliest',
    'ssl.ca.location': 'ca.pem',
    'ssl.endpoint.identification.algorithm': 'https'
}

consumer = Consumer(conf)
consumer.subscribe(['orders'])

print("Listening for live orders on Aiven securely...")

try:
    while True:
        msg = consumer.poll(timeout=1.0)
        if msg is None:
            continue
        if msg.error():
            print(f"Consumer error: {msg.error()}")
            continue
        print(f"Received message: {msg.value().decode('utf-8')}")
except KeyboardInterrupt:
    print("\nStopping consumer...")
finally:
    consumer.close()
```
